```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with 09_30 am python class.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{1,2},\s\d{1,2}:\d{1,2}\s\w{1,2}\s-\s'
```

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
type(dates)
```

```python
nl=[]
for i in dates:
    ##print(i)
    q=i.replace("\u202fam - ","-")
    r=q.replace("\u202fpm - ","-")
    ##print(r)
    nl.append(r)
print(nl)
```

```python
type(nl)
```

```python
print(nl)
```

```python
ND=[]
NT=[]
for i in nl:
    ND.append(i[0:8])
    r=i[10:]
    l=len(r)
    if(l==5):
        j=str(0)+i[10:]
        NT.append(j)
    else:
        NT.append(i[10:])
```

```python
ND
```

```python
NT
```

```python
RDT=[]
for i in range(0,len(ND)):
    RDT.append(ND[i]+","+NT[i])
RDT
```

```python
df = pd.DataFrame({'user_message':messages,'message_date':RDT})
df['message_date'] = pd.to_datetime(df['message_date'],format='%d/%m/%y,%I:%M-')
df.rename(columns={'message_date':'date'},inplace=True)
df.head()
```

```python
df.shape
```

```python
## separate users and messages
users =[]
messages =[]
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'],inplace=True)
```

```python
df.head(10)
```

```python
#Extract Year
df['year'] = df['date'].dt.year
df.head(5)
```

```python
#Extract Month
df['month'] = df['date'].dt.month_name()
df.head(5)
```

```python
#Extract Day
df['day'] = df['date'].dt.day
df.head(5)
```

```python
#Extract Hour
df['hour'] = df['date'].dt.hour
df.head(5)
```

```python
#Extract Minutes
df['minute'] = df['date'].dt.minute
df.head(5)
```
